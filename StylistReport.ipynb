```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('StylistReport.csv')
```

```python
data.shape
```

# 1. Stylist Reports

```python
data.head()
```

## 1.1 Stylist Performance

```python
plt.plot(data['StylistName'], data['CompletedAppointments'])
plt.title('Stylist Performance')
plt.xlabel('Stylist Name')
plt.ylabel('Completed Appointments')
plt.show()
```

## 1.2 Stylists Availability

```python
name = data['StylistName']
attendance = data['AttendancePercentage']

# Create a bar chart
plt.bar(name, attendance)
plt.xlabel('StylistName')
plt.ylabel('Attendance')
plt.title('Stylist Attendance Bar Chart')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

# 2. Service Reports

```python
data = pd.read_csv('ServiceReport.csv')
```

```python
data.head()
```

## 2.1 Top Revenue by Service

```python
service = data['Service']
revenue = data['Total Revenue']

# Create a bar chart
plt.bar(service, revenue, color='orange')
plt.xlabel('Service Name')
plt.ylabel('Top Revenue')
plt.title('TOP REVENUE BY SERVICE')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

## 2.2 Service wise Appointments percentage

```python
categories = data['Service']
values = data['Total Appointments']
colors = ['c', 'g', 'y', 'violet','pink', 'orange']

plt.figure(figsize=(5, 5))

# Create the pie chart
plt.pie(values, labels=categories, colors=colors,
        autopct='%1.2f%%',
        startangle=140)

plt.title('Service wise Appointments percentage')

# Display the pie chart
plt.axis('equal')
plt.show()
```

## 2.3 Service wise Average Rating

```python
plt.plot(data['Service'], data['Average Rating'])
plt.title('Service wise average ratings')
plt.xlabel('Service Name')
plt.ylabel('Ratings')
plt.show()
```

```python
data1 = data['Average Rating']
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size

# Create the histogram
plt.hist(data1, bins=20, edgecolor='k')  # You can adjust the number of bins

# Add labels and a title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the histogram
plt.show()
```
